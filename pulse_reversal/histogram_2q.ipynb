```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
Q1_list = np.load('Q1_list.npy')
Q2_list = np.load('Q2_list.npy')
```

```python
states = ("|00>", "|10>", "|01>", "|11>")
result_magnitudes = {
    'Q1': Q1_list,
    'Q2': Q2_list,
}

x = np.arange(len(states))  # the label locations
width = 0.25  # the width of the bars
multiplier = 0

fig, ax = plt.subplots(layout='constrained')

for attribute, measurement in result_magnitudes.items():
    offset = width * multiplier
    rects = ax.bar(x + offset, measurement, width, label=attribute)
    ax.bar_label(rects, padding=3)
    multiplier += 1

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Magnitude')
ax.set_title('NTU qubits baseline, amplitude = 1')
ax.set_xticks(x + width, states)
ax.legend(loc='upper left', ncols=3)

ax.set_ylim(0, 80000)
plt.show()
plt.savefig('Baseline')
```
