```python
import numpy as np
pi_pulse_duration = 250
idle_duration = 200

times = np.load('pulse_reversal_times.npy')
expects = np.load('pulse_reversal_expect.npy')
time_idle_1 = idle_duration
time_pulse = pi_pulse_duration + time_idle_1
time_idle_2 = idle_duration + time_pulse
time_pulse_2 = pi_pulse_duration + time_idle_2

import matplotlib.pyplot as plt
plt.rcParams["figure.figsize"] = (5,3)
plt.rcParams['text.usetex'] = True
plt.rcParams["font.family"] = 'serif'
plt.rcParams["figure.dpi"] = 200

plt.figure()
plt.plot(times, expects)
plt.xlabel('Time t (ns)')
plt.ylabel(f'Average expectation values of the shots\n(unknown axis scale, max = {np.round(max(expects))})')
plt.axvspan(0,time_idle_1, alpha=0.2, color='red', label = 'Always on coupling free evolution')
plt.axvspan(time_idle_1,time_pulse, alpha=0.2, color='blue', label = r'$I\otimes X$ pulse')
plt.axvspan(time_pulse,time_idle_2, alpha=0.2, color='red')
plt.axvspan(time_idle_2,time_pulse_2, alpha=0.2, color='blue')
plt.legend(bbox_to_anchor=(1.01, -0.2),shadow=True, ncol=2)
plt.title('Attempts to invert always on interaction\nusing Hahn echo with initial state 01')
plt.show()
```
